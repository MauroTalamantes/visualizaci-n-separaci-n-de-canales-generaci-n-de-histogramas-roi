```python
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/MyDrive/
# Libreras
import numpy as np
import matplotlib.pyplot as plt
import cv2
img_cv = cv2.imread('/content/drive/MyDrive/imagen.jpg')
print(type(img_cv))
print(img_cv.dtype) # Tipo de datos de los elementos de la matriz
print(img_cv.shape) # Dimensiones de la imagen
print(img_cv.size) # Nmero total de elementos en la matriz = 900 x 1200 x
3
print(img_cv.ndim) # Nmero de dimensiones de la matriz = alto, ancho,nmero de canales (RGB)
print(np.max(img_cv)) # Valor mximo en la imagen
print(np.min(img_cv)) # Valor mnimo en la imagen
plt.imshow(img_cv)
plt.show()
# OpenCV lee la imagen en BGR pero Matplotlib espera una imagen en formatoRGB
# Conversin
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# Visualizar los canales por separado
plt.figure(figsize = (18, 6))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb[:,:,0], cmap = 'gray')
plt.axis('off')
plt.title('Canal rojo (escala de grises)')
plt.subplot(1, 3, 2)
plt.imshow(img_rgb[:,:,1], cmap = 'gray')
plt.axis('off')
plt.title('Canal verde (escala de grises)')
plt.subplot(1, 3, 3)
plt.imshow(img_rgb[:,:,2], cmap = 'gray')
plt.axis('off')
plt.title('Canal azul (escala de grises)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15, 5))
# Histograma para canal rojo
plt.subplot(1, 3, 1)
plt.bar(range(256), np.histogram(img_rgb[:,:,0].flatten(),bins = 256,
range = [0, 255])[0], color = 'r', alpha = 0.7)
plt.xlabel('Valor de pxel')
plt.ylabel('Frecuencia')
plt.title('Histograma - Canal rojo')
# Histograma para canal verde
plt.subplot(1, 3, 2)
plt.bar(range(256), np.histogram(img_rgb[:,:,1].flatten(),bins = 256,
range = [0, 255])[0], color = 'g', alpha = 0.7)
plt.xlabel('Valor de pxel')
plt.ylabel('Frecuencia')
plt.title('Histograma - Canal verde')
# Histograma para canal azul
plt.subplot(1, 3, 3)
plt.bar(range(256), np.histogram(img_rgb[:,:,2].flatten(),bins = 256,
range = [0, 255])[0], color = 'b', alpha = 0.7)
plt.xlabel('Valor de pxel')
plt.ylabel('Frecuencia')
plt.title('Histograma - Canal azul')
plt.tight_layout()
plt.show()
```

```python
# Dimensiones de la imagen
h, w, _ = img_rgb.shape
# Definir la regin de inters (ROI) (cuadrado central del 50% del tamao original)
x_start = w // 4
y_start = h // 4
x_end = 3 * w // 4
y_end = 3 * h // 4
roi = img_rgb[y_start:y_end, x_start:x_end]
# Convertir la ROI a escala de grises
roi_gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
# Mostrar la regin recortada
plt.imshow(roi_gray,cmap="gray")
plt.title("Regin de Inters (ROI)")
plt.axis('off')
plt.show()
```
